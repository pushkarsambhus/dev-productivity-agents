```python
import ipdb

def group_by_length(words: list) -> dict:
    lengthOutput = {}
    similarLengthList = []
    for word in words:
        length = len(word)
        ipdb.set_trace()  # debugger pauses here each iteration
        if lengthOutput.get(length):
            existingList = lengthOutput[length]
            existingList.append(word)
            lengthOutput[length] = existingList
        else:
            similarLengthList.append(word)
            lengthOutput[length] = similarLengthList
    
    return lengthOutput
    
print(group_by_length(['cat','dog','elephant','ox','hi']))
```

```python
from openai import OpenAI
client = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")  # local, free
```

```python
r = client.chat.completions.create(
    model="llama3",  # or "llama3.2:3b" for faster replies
    messages=[{"role":"user","content":"Say hello in one short line."}],
    temperature=0.3
)
print(r.choices[0].message.content)
```

```python
import ipywidgets as w
from IPython.display import display

prompt = w.Textarea(
    value="Give me 3 QA test ideas for a vector DB ingestion pipeline.",
    layout=w.Layout(width="100%", height="120px"))
run_btn = w.Button(description="Ask LLM", button_style="primary")
out = w.Output()

def run(_):
    out.clear_output()
    with out:
        r = client.chat.completions.create(
            model="llama3",
            messages=[{"role":"user","content":prompt.value}],
            temperature=0.3
        )
        print(r.choices[0].message.content)

run_btn.on_click(run)
display(prompt, run_btn, out)
```

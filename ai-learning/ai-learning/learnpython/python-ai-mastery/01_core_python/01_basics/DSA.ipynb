```python
def compress(s: str) -> str:
    if not s:
        return s

    result = []
    count = 1

    for i in range(1, len(s)):
        if s[i] == s[i-1]:      # same as previous
            count += 1
        else:                    # different — flush current run
            result.append(s[i-1] + str(count))
            count = 1

    # Flush the last run
    result.append(s[-1] + str(count))

    compressed = "".join(result)
```

```python
def wordsWithSubstring(words: list, sub: str) -> list:
    result = []
    for item in words:
        if sub in item:
            result.append(item)
    
    return result

wordsWithSubstring(["ant", "banana", "plant", "tree", "elephant"], "ant")
```

```python

PATTERN = "aeiouAEIOU"

def countVowels(s:str)->int:
    return sum( 1 for char in s if char in PATTERN)

countVowels("Hello World")
```

```python
def reverseString(s:str) -> str:
    reversed_s = s.split()
    reversed_s = reversed_s[::-1]
    return " ".join(reversed_s)


print(reverseString("Hello World"))
```

```python
from collections import Counter

def isAnagram(s : str , t : str) -> bool:
   return Counter(s) == Counter(t)


print(isAnagram("listen", "silent"))   # True
print(isAnagram("hello", "world"))     # False
print(isAnagram("rat", "car"))         # False

```

```python
class Solution:
    def countChar(self, s : str) -> str:

        charMap = {}

        for char in s:
            charMap[char] = charMap.get(char , 0) + 1


        for key,val in charMap.items():
            if val == 1:
                return s.index(key)
        

        return -1
```

```python
class Solution:
    
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
        
        freq = {}
        
        for char in s:
            freq[char] = freq.get(char, 0) + 1
        
        for char in t:
            if char not in freq:
                return False
            freq[char] -= 1
            if freq[char] < 0:
                return False
        
        return True
```

```python
class FizzBuzz:

    def isFizzBuzz(self , number : int):
        if number % 3 == 0 and number % 5 == 0:
            print("FizzBuzz")
        elif number % 3 == 0:
            print("Fizz")
        elif number % 5 == 0:
            print("Buzz")
        else:
            print(number)


n = int(input())
fb = FizzBuzz()
for i in range(1,n+1):
    number = int(input())
    fb.isFizzBuzz(number)
```

```python
# Log Rate Limiter

log_map = {}

def should_print_log(level: str, timestamp: int) -> bool:
    # your code here
    if level in log_map:
        if timestamp - log_map[level] >= 5:
            log_map[level] = timestamp
            return True
        else:
            return False
    else:
        log_map[level] = timestamp
        return True

logs = [
  ("INFO", 1),
  ("ERROR", 2),
  ("ERROR", 3),
  ("INFO", 8),
  ("ERROR", 10)
]

for level, timestamp in logs:
    print(should_print_log(level, timestamp))
```

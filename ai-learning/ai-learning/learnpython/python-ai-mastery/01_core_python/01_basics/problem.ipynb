```python
# Enter your code here. Read input from STDIN. Print output to STDOUT

import re


def replaceString(string):
    replacedString = re.sub(r"(?<= )&&(?= )" , " and " , string)
    return re.sub(r"(?<= )\|\|(?= )", " or ", replacedString)


numberoflines = int(input())
for i in range(numberoflines):
    string = input()
    print(replaceString(string))
```

```python
import re

PATTERN = r"(?<=[^aeiouAEIOU])[aeiouAEIOU]{2,}(?=[^aeiouAEIOU])"

def countVowels(string):
    
    results = re.findall(PATTERN, string)

    if results:
        for i in results:
            print (i)
    else:
        print ("-1")

string = input()
countVowels(string)
```

```python
import re


PATTERN = r"^[+-]?[0-9]*/.[0-9]+$"

def isFloatingPoint(number):
    if re.match(PATTERN,number):
        return True

    return False

countOfNumbers = int(input())
for i in range(countOfNumbers):
    number = input()
    print(isFloatingPoint(number))
```

```python
import re
import email.utils

PATTERN = r"[a-zA-Z][\w.-]*@[a-zA-Z]+\.[a-zA-Z]{1,3}$"

def isvalidEmail(address):
    name = email.utils.parseaddr(address)[0]
    emailaddress = email.utils.parseaddr(address)[1]

    if re.match(PATTERN,emailaddress):
        print(f"{name} <{emailaddress}")
    
numberofEmails = int(input())
for i in range(numberofEmails):
    address = input()
    isvalidEmail(address)







```

```python
# ── Exercise 2: Conditionals ───────────────────────────────────────────────────
# TODO: Write a function that classifies an engineer's seniority.
# Rules: 0-2 yrs → "junior", 3-5 → "mid", 6-10 → "senior", 11+ → "principal"

def classify_engineer(years_exp: int) -> str:
    if years_exp <=2:
        return "junior"
    if years_exp <= 5:
        return "mid"
    if years_exp <=10:
        return "senior"
    return "principal"

classify_engineer(10)
```
